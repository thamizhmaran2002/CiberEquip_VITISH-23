```python
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
img = cv2.imread("/kaggle/input/sample1234/sample5.png")

new_height = 200

# Create a black image with the new height
img1 = np.zeros((new_height, img.shape[1], 3), dtype=np.uint8)

# Place the original image on top of the black image
img1[:img.shape[0], :img.shape[1]] = img

# Define the tilt and shear transformation matrix
M = np.float32([[1, 0.5, 0], [0.5, 1, 0]])

# Perform the tilt and shear transformation
img_transformed = cv2.warpAffine(img1, M, (img1.shape[1], img1.shape[0]))

# Define the inverse tilt and shear transformation matrix
M_inv = np.linalg.inv(np.vstack([M, [0, 0, 1]]))[:2]

# Perform the inverse tilt and shear transformation
img_back = cv2.warpAffine(img_transformed, M_inv, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(img_transformed)
```

```python
plt.imshow(img_back)
```
